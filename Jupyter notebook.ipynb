### Vamos a aproximar el valor de pi utilizando la fórmula de Leibniz.

```python
def approx_pi(num_terms):
    """Aproxima el valor de pi utilizando la fórmula de Leibniz.

    Args:
        num_terms (int): Número de términos de la serie que se utilizarán en la aproximación.

    Returns:
        float: Aproximación del valor de pi.
    """
    approx = 0
    sign = 1

    for i in range(num_terms):
        approx += sign / (2*i + 1)
        sign = -sign

    return 4 * approx

# Llamamos a la función para aproximar pi utilizando 1000 términos
approximation = approx_pi(1000)

# Imprimimos la aproximación
print("Aproximación de pi:", approximation)
```
